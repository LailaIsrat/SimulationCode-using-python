```python
import numpy as np
import math
```

```python
Ax=[10]
Ay=[0]

Bx=[0]
By=[10]

Cx=[10]
Cy=[10]

Dx=[0]
Dy=[0]

Av=3
Bv=5
Cv=7
Dv=2

shoot_A =0
shoot_B = 0
shoot_C = 0
shoot_D = 0

is_shooting_value = False
DA_distence=math.sqrt((Dx[0]-Ax[0])**2 + (Dy[0]-Ay[0])**2)
CB_distence=math.sqrt((Cx[0]-Bx[0])**2 + (Cy[0]-By[0])**2)
BD_distence=math.sqrt((Bx[0]-Dx[0])**2 + (By[0]-Dy[0])**2)

for t in range (1,21):
  print()
  print("Timing iteration = ",t)
 
  #D chased by A : target D chased by A

  sinTheta= (Dy[t-1] - Ay[t-1]) / DA_distence
  cosTheta = (Dx[t-1] - Ax[t-1]) / DA_distence
  newAx=Ax[t-1] + (Av*sinTheta)
  newAy=Ay[t-1] + (Av*cosTheta)
  Ax.append(newAx)
  Ay.append(newAy)

  #C is chased by B ,Target C chased by B

  sinTheta= (Cy[t-1] - By[t-1]) / CB_distence
  cosTheta = (Cx[t-1] - Bx[t-1]) / CB_distence
  newBx=Bx[t-1] + (Bv*sinTheta)
  newBy=By[t-1] + (Bv*cosTheta)
  Bx.append(newBx)
  By.append(newBy)

  #B is chased by D ,Target B  chased by D

  sinTheta= (By[t-1] - Dy[t-1]) / BD_distence
  cosTheta = (Bx[t-1] - Dx[t-1]) / BD_distence
  newDx=Dx[t-1] + (Dv*sinTheta)
  newDy=Dy[t-1] + (Dv*cosTheta)
  Dx.append(newDx)
  Dy.append(newDy)


  
  newCx=Cx[t-1] 
  newCy=Cy[t-1] + Cv

  Cx.append(newCx)
  Cy.append(newCy)
  
  print("x and y cordinate for car A =(",Ax[t] ,",",Ay[t],")")
  print("x and y cordinate for car B =(",Bx[t],",",By[t],")")
  print("x and y cordinate for car C =(",Cx[t],",",Cy[t],")")
  print("x and y cordinate for car D =(",Dx[t],",",Dy[t],")")

  DA_distence=math.sqrt((Dx[t]-Cx[t])**2 + (Dy[t]-Cy[t])**2)
  CB_distence=math.sqrt((Cx[t]-Bx[t])**2 + (Cy[t]-By[t])**2)
  BD_distence=math.sqrt((Bx[t]-Dx[t])**2 + (By[t]-Dy[t])**2)

 
  if DA_distence < 5:
    is_shooting_value == True
    print("A shooting D at time=",t,"Distence between DA",DA_distence)
    shoot_D +=1
    
  if CB_distence < 5:
    is_shooting_value == True
    print("B shooting C at time=",t,"Distence between CB",CB_distence)
    shoot_C +=1
  if BD_distence < 5:
    is_shooting_value == True
    print("D shooting B at time=",t,"Distence between BD",BD_distence)
    shoot_B +=1
 # if is_shooting_value == False :
    ##print()
    #print("No shooting occured in this time = ",t)
    #break
print()
print("Car A is shooted = ",shoot_A,"times")
print("Car B is shooted = ",shoot_B,"times")
print("Car C is shooted = ",shoot_C,"times")
print("Car D is shooted = ",shoot_D,"times")
```

```python
import matplotlib.pyplot as plt

plt.plot(Ax,Ay,label = "car A")
plt.plot(Bx,By,label = "car B")
plt.plot(Cx,Cy,label = "car C")
plt.plot(Dx,Dy,label = "car D")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.legend()
plt.show()
```
