```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x_A=[10]
y_A=[0]

x_B=[0]
y_B=[10]

x_C=[10]
y_C=[10]

x_D=[0]
y_D=[0]

v_A=3
v_B=5
v_C=7
v_D=2
#time =20

count_A=0
count_B=0
count_C=0
count_D=0
Distence_DB  = math.sqrt((x_D[0]-x_B[0])**2 + (y_D[0]-y_B[0])**2)
Distence_CA = math.sqrt((x_C[0]-x_A[0])**2 + (y_C[0]-y_A[0])**2)
Distence_BC  = math.sqrt((x_B[0]-x_C[0])**2 + (y_B[0]-y_C[0])**2)
Distence_AD  = math.sqrt((x_A[0]-x_D[0])**2 + (y_A[0]-y_D[0])**2)

for t in range (1,21):
  #D chased by B : target D chased by B
  print()
  print("for time = ",t)
  sin_theta= (y_D[t-1] - y_B[t-1]) / Distence_DB
  cos_theta = (x_D[t-1] - x_B[t-1]) / Distence_DB
  x_B_new=x_B[t-1] + (v_B*cos_theta)
  y_B_new=x_B[t-1] + (v_B*sin_theta)
  x_B.append(x_B_new)
  y_B.append(y_B_new)

  #c is chased by A ,Target C defender/chaser A

  sin_theta= (y_C[t-1] - y_A[t-1]) / Distence_CA
  cos_theta = (x_C[t-1] - x_A[t-1]) / Distence_CA
  x_A_new=x_A[t-1] + (v_A*cos_theta)
  y_A_new=y_A[t-1] + (v_A*sin_theta)
  x_A.append(x_A_new)
  y_A.append(y_A_new)

  #B is chased by C ,Target B  defender/chaser C
  sin_theta= (y_B[t-1] - y_C[t-1]) / Distence_BC
  cos_theta = (x_B[t-1] - x_C[t-1]) / Distence_BC
  x_C_new=x_C[t-1] + (v_C*cos_theta)
  y_C_new=y_C[t-1] + (v_C*sin_theta)
  x_C.append(x_C_new)
  y_C.append(y_C_new)

 #A is chased by D ,Target A  defender/chaser D
  sin_theta= (y_A[t-1] - y_D[t-1]) / Distence_AD
  cos_theta = (x_A[t-1] - x_D[t-1]) / Distence_AD
  x_D_new=x_D[t-1] + (v_D*cos_theta)
  y_D_new=y_D[t-1] + (v_D*sin_theta)

  x_D.append(x_D_new)
  y_D.append(y_D_new)

  print()

  #print("x and y cordinate value of car A =(","%.2f"%x_A[t] "," "%.2f"%y_A[t]") at time = ",t)
  
  #print("Formatted Number: "+"{:.2f}".format(x));

  #print("x and y cordinate value of car A =("+"{:.2f}".format(x_A[t]) ,"," +"{:.2f}".format(y_A[t]),") at time = ",t)
  #print("x and y cordinate value of car B =("+"{:.2f}".format(x_B[t]),"," +"{:.2f}".format(y_B[t]),") at time = ",t)
  #print("x and y cordinate value of car C =("+"{:.2f}".format(x_C[t]),","+"{:.2f}".format(y_C[t]),") at time = ",t)
  #print("x and y cordinate value of car D =("+"{:.2f}".format(x_D[t]),","+"{:.2f}".format(y_D[t]),")  at time = ",t)
  print("x and y cordinate value of car A =(",x_A[t],",",y_A[t],") at time = ",t)
  print("x and y cordinate value of car B =(",x_B[t],",",y_B[t],") at time = ",t)
  print("x and y cordinate value of car C =(",x_C[t],",",y_C[t],") at time = ",t)
  print("x and y cordinate value of car D =(",x_D[t],",",y_D[t],") at time = ",t)
  Distence_DB  = math.sqrt((x_D[t]-x_B[t])**2 + (y_D[t]-y_B[t])**2)
  Distence_CA  = math.sqrt((x_C[t]-x_A[t])**2 + (y_C[t]-y_A[t])**2)
  Distence_BC  = math.sqrt((x_B[t]-x_C[t])**2 + (y_B[t]-y_C[t])**2)
  Distence_AD  = math.sqrt((x_A[t]-x_D[t])**2 + (y_A[t]-y_D[t])**2)
  print()
  if Distence_DB < 5:
    #print()
    print("B Shooting D at time = ",t,"Distence between DC = ",Distence_DB)
    count_D +=1
    #break
  if Distence_CA <5 :
    #print()
    print("A shooting C at time = ",t,"Distence between CA = ",Distence_CA)
    count_C +=1
    #break
  if Distence_BC < 5:
    #print()
    print("A shooting B at time = ",t,"Distence between BC = ",Distence_BC)
    count_B +=1
    #break
  if Distence_AD < 5:
    #print()
    print("D shooting A at time = ",t,"Distence between AD = ",Distence_AD)
    count_A +=1
    #break
  
print()
print("Target X value of A car",x_A)
print("Target Y value of A car ",y_A)

print()
print("Defender X value of B car",x_B)
print("Defender Y value of B car",y_B)

print()
print("Defender X value of C car",x_C)
print("Defender Y value of C car",y_C)

print()
print("Defender X value of D car",x_D)
print("Defender Y value of D car",y_D)

print()
print()
print("Car A is shooted = ",count_A,"times")
print("Car B is shooted = ",count_B,"times")
print("Car C is shooted = ",count_C,"times")
print("Car D is shooted = ",count_D,"times")




#********************************graph***********************


plt.plot(x_A,y_A,label = "car A")
plt.plot(x_B,y_B,label = "car B")
plt.plot(x_C,y_C,label = "car C")
plt.plot(x_D,y_D,label = "car D")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.legend()
plt.show()
```
