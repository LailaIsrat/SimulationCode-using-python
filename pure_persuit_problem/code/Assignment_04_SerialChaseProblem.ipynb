```python
import numpy as np
import math
```

```python
Ax=[10]
Ay=[0]

Bx=[0]
By=[10]

Cx=[10]
Cy=[10]

Dx=[0]
Dy=[0]

Av=3
Bv=5
Cv=7
Dv=2

shoot_A =0
shoot_B = 0
shoot_C = 0
shoot_D = 0

is_shooting_value = False
DB_distence=math.sqrt((Dx[0]-Bx[0])**2 + (Dy[0]-By[0])**2)
CA_distence=math.sqrt((Cx[0]-Ax[0])**2 + (Cy[0]-Ay[0])**2)
BC_distence=math.sqrt((Bx[0]-Cx[0])**2 + (By[0]-Cy[0])**2)
AD_distence=math.sqrt((Ax[0]-Dx[0])**2 + (Ay[0]-Dy[0])**2)


for t in range (1,21):
  
  print()
  print("Timing iteration = ",t)
  #D chased by B : target D chased by B

  sinTheta= (Dy[t-1] - By[t-1]) / DB_distence
  cosTheta = (Dx[t-1] - Bx[t-1]) / DB_distence
  newBx=Bx[t-1] + (Bv*sinTheta)
  newBy=By[t-1] + (Bv*cosTheta)
  Bx.append(newBx)
  By.append(newBy)

  #C is chased by A ,Target C defender/chaser A

  sinTheta= (Cy[t-1] - Ay[t-1]) / CA_distence
  cosTheta = (Cx[t-1] - Ax[t-1]) / CA_distence
  newAx=Ax[t-1] + (Av*sinTheta)
  newAy=Ay[t-1] + (Av*cosTheta)
  Ax.append(newAx)
  Ay.append(newAy)

  #B is chased by C ,Target B  defender/chaser C

  sinTheta= (By[t-1] - Cy[t-1]) / BC_distence
  cosTheta = (Bx[t-1] - Cx[t-1]) / BC_distence
  newCx=Cx[t-1] + (Cv*sinTheta)
  newCy=Cy[t-1] + (Cv*cosTheta)
  Cx.append(newCx)
  Cy.append(newCy)


  #A chased by D : target A chased by D

  sinTheta= (Dy[t-1] - Ay[t-1]) / AD_distence
  cosTheta = (Dx[t-1] - Ax[t-1]) / AD_distence
  newDx=Dx[t-1] + (Dv*sinTheta)
  newDy=Dx[t-1] + (Dv*cosTheta)
  Dx.append(newDx)
  Dy.append(newDy)


  print("x and y cordinate for car A =(",Ax[t],",",Ay[t],")")
  print("x and y cordinate for car B =(",Bx[t],",",By[t],")")
  print("x and y cordinate for car C =(",Cx[t],",",Cy[t],")")
  print("x and y cordinate for car D =(",Dx[t],",",Dy[t],")")

  DB_distence=math.sqrt((Dx[t]-Bx[t])**2 + (Dy[t]-By[t])**2)
  CA_distence=math.sqrt((Cx[t]-Ax[t])**2 + (Cy[t]-Ay[t])**2)
  BC_distence=math.sqrt((Bx[t]-Cx[t])**2 + (By[t]-Cy[t])**2)
  AD_distence=math.sqrt((Ax[t]-Dx[t])**2 + (Ay[t]-Dy[t])**2)


  if DB_distence < 5:
    is_shooting_value == True
    print("B shooting D at time=",t,"Distence between DB",DB_distence)
    shoot_B +=1
    
  if CA_distence < 5:
    is_shooting_value == True
    print("A shooting C at time=",t,"Distence between CA",CA_distence)
    shoot_A +=1
  if BC_distence < 5:
    is_shooting_value == True
    print("C shooting B at time=",t,"Distence between BC",BC_distence)
    shoot_C +=1

  if AD_distence < 5:
    is_shooting_value == True
    print(" D shooting A at time=",t,"Distence between AD",AD_distence)
    shoot_D +=1

print()
print("Car A is shooted = ",shoot_A,"times")
print("Car B is shooted = ",shoot_B,"times")
print("Car C is shooted = ",shoot_C,"times")
print("Car D is shooted = ",shoot_D,"times")
```

```python
import matplotlib.pyplot as plt

plt.plot(Ax,Ay,label = "car A")
plt.plot(Bx,By,label = "car B")
plt.plot(Cx,Cy,label = "car C")
plt.plot(Dx,Dy,label = "car D")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.legend()
plt.show()
```
